# Better states data using the 'us' library

```python
import pandas as pd
import geopandas as gpd
import us
```

## States metadata

```python
all_features = []

for st in us.states.STATES:
    states_dict = {
        'fips': st.fips,
        'name': st.name,
        'abbr': st.abbr,
        "is_continental": st.is_continental,
        "statehood_year": st.statehood_year,
        "capital": st.capital,
        "capital_tz": st.capital_tz,
        "ap_abbr": st.ap_abbr,
        "shapefile_urls": st.shapefile_urls()
    }
    all_features.append(states_dict)
```

```python
df = pd.DataFrame(all_features).sort_values('fips', ascending=True).reset_index(drop=True)
```
