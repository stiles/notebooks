```python
import requests
import pandas as pd
from random import randint
from time import sleep
from tqdm.notebook import tqdm
```

```python
season_dfs = []

for year in tqdm(range(1994,2024)):
    url = f'https://www.pro-football-reference.com/years/{year}/penalties.htm'
    season_df = pd.read_html(url)[0]
    season_df['season'] = year
    season_df.columns = ['penalty_name', 'penalty_count', 'yards', 'penalties_per_game', 'yards_per_game', 'season']
    season_dfs.append(season_df)
    sleep(randint(2, 6))
```

```python
df = pd.concat(season_dfs).query('season >= 1999')
```
