```python
mi_candidates["statement year"] = pd.to_datetime(
    mi_candidates["statement year"], format="%Y"
)

mi_candidates["report covers to"] = pd.to_datetime(
    mi_candidates["report covers to"], format="%m/%d/%y"
)
```

```python
years = ["2019-01-01", "2020-01-01", "2021-01-01", "2022-01-01"]
```

```python
mi_candidates = mi_candidates[mi_candidates["statement year"].isin(years)]
```

```python
mi_candidates[mi_candidates["report covers to"] > "2019-01-01"]
```

```python
mi_candidates_clean = mi_candidates[
    [
        "committee id#",
        "document sequence#",
        "committee name",
        "candidate first name",
        "candidate middle name",
        "candidate last name",
        "political party affiliation",
        "office sought",
        "district sought",
        "statement year",
        "date filed",
        "report covers from",
        "report covers to",
        "total contributions/other receipts cumulative this election cycle",
        "inkind contributions cumulative this election cycle",
        "inkind expenditures cumulative this election cycle",
        "total expenditures cumulative this election cycle",
        "ending balance of this report",
    ]
]
```

```python
mi_candidates_clean.columns = mi_candidates_clean.columns.str.replace(
    " ", "_"
).str.replace("#", "")
```

```python
mi_candidates_clean["report_covers_from"] = pd.to_datetime(
    mi_candidates_clean["report_covers_from"], format="%m/%d/%y"
)
mi_candidates_clean["date_filed"] = pd.to_datetime(
    mi_candidates_clean["date_filed"], format="%m/%d/%y"
)
```

```python
mi_candidates_clean.dtypes
```

```python
mi_candidates_clean.candidate_first_name = (
    mi_candidates_clean.candidate_first_name.str.title()
)
mi_candidates_clean.candidate_last_name = (
    mi_candidates_clean.candidate_last_name.str.title()
)
```

```python
mi_candidates_clean["candidate"] = mi_candidates_clean.candidate_first_name.str.cat(
    mi_candidates_clean.candidate_last_name, sep=" "
)
```

```python
mi_q_candidates = state_candidates[state_candidates.state == "MI"]
```

```python
qmich_names = mi_q_candidates["candidate"]
```

```python
mi_q_campaign_finance = mi_candidates_clean[
    mi_candidates_clean.candidate.isin(qmich_names)
]
```

```python
mi_q_campaign_finance = mi_q_campaign_finance[
    [
        "committee_id",
        "candidate",
        "report_covers_from",
        "report_covers_to",
        "total_contributions/other_receipts_cumulative_this_election_cycle",
        "total_expenditures_cumulative_this_election_cycle",
        "ending_balance_of_this_report",
    ]
]
```

```python
mi_q_campaign_finance_2022 = mi_q_campaign_finance[
    mi_q_campaign_finance.report_covers_to > "2021-12-30"
]
```

```python
mi_q_summary = mi_q_campaign_finance_2022.groupby(["candidate", "committee_id"]).agg(
    receipts=("total_contributions/other_receipts_cumulative_this_election_cycle", sum),
    spending=("total_expenditures_cumulative_this_election_cycle", sum),
    cash=("ending_balance_of_this_report", sum),
)
```
