### get team links

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup


html = urlopen("https://en.wikipedia.org/wiki/National_Women%27s_Soccer_League")
```

```python
soup = BeautifulSoup(html, "html.parser")
table = soup.find_all("tbody")[1]
rows = table.find_all("tr")
```

```python
link_stems = []
for row in rows[1:]:
    rowlinks = row.find_all("a")
    link_stems.append(rowlinks[0].get("href"))
```

```python
link_stems
```

```python
links = []
for link in link_stems:
    links.append("https://en.wikipedia.org/" + str(link))
```

```python
team_names = []
coach = []
for row in rows[1:]:
    cells = row.find_all("td")
    team_names.append(cells[0].text)
    coach.append(cells[6].text)
```

```python
# for now just put what you have together in a df
df = pd.DataFrame({"team": team_names, "coach": coach, "wikipedia": links})
```
