# U.S. Counties - Merge processed FIPS codes to geo files

```python
import pandas as pd
import jenkspy 
import geopandas as gpd
from pathlib import Path
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import json
import os
import seaborn as sns
import altair_latimes as lat
import altair as alt
from pandas.io.json import json_normalize 
alt.renderers.enable('notebook')
alt.themes.register('latimes', lat.theme)
alt.themes.enable('latimes')
```

```python
# Codes via U.S. Census Bureau. Related demographics via Esri. 
src = pd.read_excel('input/fips.xlsx')
```

```python
src.columns = src.columns.str.strip().str.lower().str.replace(' ', '_')\
                    .str.replace('(', '').str.replace(')', '').str.replace('-','_')
```

```python
src.loc[0]
```

```python
codes = src.astype({'objectid': 'str',\
                        'state_fips': 'str',\
                        'cnty_fips': 'str',\
                        'fips': 'str'}, inplace=True)
```

```python
codes.head()
```

```python
fips_counties = pd.DataFrame(codes[['fips','name','state_name','pop2010']])
```

```python
fips_counties.dtypes
```

```python
fips_counties = fips_counties.rename({'fips':'geoid', 'name': 'county', 'state_name': 'state', 'pop2010':'population'}, axis='columns')
```
